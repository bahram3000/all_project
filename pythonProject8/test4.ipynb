```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def history_average(data_):
    l = []
    for i in range(1, len(data_) + 1):
        l.append(np.mean(data_[:i]))
    #l.pop(0)
    return np.array(l)
```

```python
data = np.random.random(100)
data
```

```python
avg = history_average(data)
avg
```

```python
plt.plot(data, c='blue')
plt.plot(avg, c='red')
```
