```python
def to_english(name:str):
    stp='ابپتثجچحخدذرزسشصضطظعغفقکگلمنوهیآ'
    dict_name={stp[0]:'i',stp[1]:'b',stp[2]:'p',stp[3]:'t',stp[4]:'c',stp[5]:'j',stp[6]:'ch',stp[7]:'h',
               stp[8]:'kh',stp[9]:'d',stp[10]:'dh',stp[11]:'r',stp[12]:'z',stp[13]:'s',stp[14]:'sh',
               stp[15]:'S',stp[16]:'SZ',stp[17]:'T',stp[18]:'TH',stp[19]:'AE',stp[20]:'GH',stp[21]:'f',stp[22]:'gh',stp[23]:'k',stp[24]:'g',stp[25]:'l',stp[26]:'m',stp[27]:'n',stp[28]:'v',stp[29]:'he',stp[30]:'y',stp[31]:'a',' ':' '}

    def to_english2(name2):
        if type(name2) == type(''):
            name_eng=''
            for i in name2:
                name_eng+=dict_name[i]
            return name_eng
    if type(name) == type(''):
        #print('str')
        return to_english2(name)
    elif type(name) == pd.core.indexes.base.Index:
        #print('list')
        list_name=[]
        for i in name:
            list_name.append(to_english2(i))
        return list_name
    else:
        return 'Error'

```
