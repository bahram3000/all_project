```python
import pickle
from sklearn.model_selection import train_test_split
```

```python
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras.models import Sequential
```

```python
with open('F:\my project\my project\\neural\digikala_silver_product\\finally_df_con_5.pickle','rb') as f:
    data=pickle.load(f)
```

```python
data
```

```python
x_data=data.iloc[:,:-1].values
y_data=data.iloc[:,-1].values
```

```python
x_data=x_data/255
x_data
```

```python
y_data
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_data,y_data,train_size=0.9,shuffle=True)
```

```python
model=Sequential([Dense(300,activation='relu',input_shape=(x_train.shape[1],)),Dense(150,activation='relu'),
                  Dense(1)])
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(x_train, y_train, epochs=2000, batch_size=73)
```

```python
pred_test=model.predict(x_test)
pred_test
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(y_test)
plt.plot(pred_test)
```
