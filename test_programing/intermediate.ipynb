```python
def number_to_english(str_num:str):
    if type(str_num)==str:
        n_d={'':'1','':'2','':'3','':'4','':'5','':'6','':'7','':'8','':'9','':'0','':'','':'.'}
        s=''
        for i in str_num:
            s+=n_d[i]
        return float(s)
```

```python
num=data_final['population']['']
num
```

```python
number_to_english(num)
```

```python
for i in data_final.columns:
    for j in data_final.index:
        if type(data_final[i][j]) != float and type(data_final[i][j]) !=int:
            data_final[i][j]=number_to_english(data_final[i][j])
```
