```python
list_image_url = []
list_rate = []
for i in range(200):
    api_url_v = f'https://api.digikala.com/v1/categories/women-silver-jewelry/search/?seo_url=&page={i}'
    param_v = f'seo_url=&page={i}'
    try:
        res_v = requests.get(url=api_url_v, params=param_v)
        data_v = res_v.json()
        print(f' {i}-done ', end='')
        for j in range(20):
            list_image_url.append(data_v['data']['products'][j]['images']['main']['url'][0])
            list_rate.append(data_v['data']['products'][j]['rating'])
    except:
        print(f' {i}-Fail ')
    time.sleep(1)
```

```python
import pandas as pd
```

```python
list_image_url = pd.Series(list_image_url)
list_image_url.drop_duplicates(inplace=True)
list_image_url
```

```python
k = 0
ind = []
for i in list_image_url:
    try:
        r_p = requests.get(i).content
        with open(f'F:\\my project\\my project\\neural\\digikala_silver_product\\img{k}.jpg', 'wb') as f:
            f.write(r_p)
        ind.append(k)
    except:
        print('fail', end=' ')
    k += 1
```

```python
df = {'rate': [], 'count': []}
for i in ind:
    df['rate'].append(list_rate[i]['rate'])
    df['count'].append(list_rate[i]['count'])
```
