```python
return_df = pd.read_excel('F:\my project\dataset\iran_index2.xlsx')
return_df.rename(index=return_df['Unnamed: 0'], inplace=True)
return_df.drop('Unnamed: 0', axis=1, inplace=True)
ret_df = return_df.replace(np.nan, 0)
return_df
```

```python
s = 0
c = 0
for i in ret_df.index:
    if ret_df['hosing_rate'][i] != 0:
        s += ret_df['hosing_rate'][i]
        c += 1
c += 1
s / c
```

```python
def mean_exp(data):
    s = 1
    n = 0
    for i in data:
        if i != 0:
            s*=i
            n+=1
    return s**(1/n)

```

```python
data_f=pd.DataFrame(data_f)
```

```python
data_f.reset_index(inplace=True)
data_f
```

```python
x=data_f.index
y=data_f['s/all'].values
```

```python
x=np.array(x).reshape((-1,1))
y=y.reshape((-1,1))
```

```python
from sklearn.linear_model import LinearRegression
lg=LinearRegression()
lg.fit(x,y)
y_p=lg.predict(x)
plt.plot(x,y)
plt.plot(x,y_p)
plt.grid()
```
