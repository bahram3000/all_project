```python
url = "https://api.divar.ir/v8/web-search/30/plot-old"
js = {"page": 0, "json_schema": {"category": {"value": "plot-old"}, "query": "زمین", "sort": {"value": "sort_date"}},
      "last-post-date": int(str(time.time()).replace('.', ''))}
```

```python
try:
    res = requests.post(url=url, json=js)
    data = res.json()
    print('done')
except:
    print(requests.post(url=url, json=js).status_code)
```
