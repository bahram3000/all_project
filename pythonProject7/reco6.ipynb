```python
from sklearn.datasets import fetch_california_housing
```

```python
data_r = fetch_california_housing()
```

```python
data_r
```

```python
x_r = data_r.data
y_r = data_r.target
```

```python
x_r
```

```python
y_r
```

```python
from sklearn.linear_model import ElasticNet
```

```python
elastic_reg = ElasticNet()
```

```python
elastic_reg.fit(x_r[:20000, :], y_r[:20000])
```

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import VotingRegressor
from sklearn.linear_model import LinearRegression
```

```python
x_t, y_t = x_r[:20000, :], y_r[:20000]
```

```python
lr = LinearRegression()
lr.fit(x_t, y_t)
gr = GradientBoostingRegressor()
gr.fit(x_t, y_t)
fr = RandomForestRegressor()
fr.fit(x_t, y_t)
```

```python
vr = VotingRegressor(estimators=[('lr', lr), ('gr', gr), ('fr', fr)])
vr.fit(x_t, y_t)
```

```python
plt.scatter(vr.predict(x_r[20000:, :]), y_r[20000:])
```

```python
from sklearn.datasets import load_iris
```

```python
x = load_iris().data
y = load_iris().target
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_tr, x_ts, y_tr, y_ts = train_test_split(x, y, train_size=0.7, shuffle=True)
```

```python
x_tr
```

```python
y_tr
```

```python
log_reg.fit(x_tr, y_tr)
```

```python
log_reg.coef_
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.scatter(x[:, 2], x[:, 3], c=y)
plt.subplot(1, 2, 2)
plt.scatter(x[:, 2], x[:, 3], c=log_reg.predict(x))
```

```python
log_reg.score(x, y)
```

```python
log_reg.coef_ @ x[0, :]
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(x_tr, y_tr)
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.scatter(x[:, 2], x[:, 3], c=y)
plt.subplot(1, 2, 2)
plt.scatter(x[:, 2], x[:, 3], c=knn.predict(x))
```

```python
knn.score(x_ts, y_ts)
```
