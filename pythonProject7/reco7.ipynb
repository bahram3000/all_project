```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
x = load_iris().data
y = load_iris().target
```

```python
plt.scatter(x[:, 2], x[:, 3], c=y)
```

```python
from sklearn.cluster import KMeans
```

```python
kmn_cluster = KMeans(n_clusters=3)
```

```python
kmn_cluster.fit(x)
```

```python
kmn_cluster.labels_
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.scatter(x[:, 2], x[:, 3], c=y)
plt.subplot(1, 2, 2)
plt.scatter(x[:, 2], x[:, 3], c=kmn_cluster.labels_)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
samples = np.array([[1, 1], [2, 2], [4, 3], [5, 3], [6, 3], [9, 6]])
plt.scatter(samples[:, 0], samples[:, 1])
```

```python
agg_cluster = AgglomerativeClustering(n_clusters=3).fit(samples)
```

```python
agg_cluster.labels_
```

```python
agg_cluster.compute_full_tree
```

```python
plt.scatter(samples[:, 0], samples[:, 1], c=agg_cluster.labels_)
```

```python
from sklearn.datasets import make_blobs
```

```python
x1, y1 = make_blobs(n_samples=200, n_features=5, centers=5)
```

```python
plt.scatter(x1[:, 0], x1[:, 1])
```

```python
from sklearn.mixture import GaussianMixture
```

```python
GMM = GaussianMixture(n_components=5)
```

```python
GMM.fit(x1, y1)
```

```python
plt.figure()
plt.subplot(1, 2, 1)
plt.scatter(x1[:, 0], x1[:, 1], c=y1)
plt.subplot(1, 2, 2)
plt.scatter(x1[:, 0], x1[:, 1], c=GMM.predict(x1))
```

```python
GMM.predict_proba(x1)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(x)
```

```python
x_pca = pca.transform(x)
```

```python
x_pca
```

```python
plt.scatter(x_pca[:, 0], x[:, 1], c=y)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
```

```python
lda.fit(x, y)
```

```python
x_lda = lda.transform(x)
```

```python
plt.scatter(x_lda[:, 0], x[:, 1], c=y)
```
