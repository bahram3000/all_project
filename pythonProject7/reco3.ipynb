```python
tweets = []
#lens=[]
for i in dir:
    path_ = path + "\\" + i
    file = open(path_, mode='r', encoding='utf-8')
    while True:
        a = file.readline().split('|')[2].strip().lower()
        tweets.append(a[:a.find('http://') - 1])
        #lens.append(len(a[:a.find('http://') - 1]))
        if file.readline() == "":
            file.close()
            break
```

```python
l1 = []
for i in range(max_l):
    l2 = []
    for j in words:
        if i < len(j):
            l2.append(j[i])
    l1.append(l2)
```

```python
l1
```

```python
dic_org = {}
for k, i in enumerate(l1):
    dic = {}
    for j in i:
        dic[j] = i.count(j)
    dic_org[k] = dic
dic_org
```
