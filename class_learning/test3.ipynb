```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'F:\my project\dataset\my pro-datasets\\maskan.xlsx'
data_df = pd.read_excel(path)
data_df
```

```python
data_df.drop('متوسط قيمت يك متر مربع زمين کلنگی(هزارريال)', axis=1, inplace=True)
data_df
```

```python
data_df.replace('-', np.nan, inplace=True)
data_df
```

```python
data_perity = pd.DataFrame(np.array([np.nan] * len(data_df.index)), index=data_df.index)
data_perity
```

```python

for i, j in enumerate(data_df.columns):
    for k, l in enumerate(data_df.columns):
        if i != k:
            if k > i:
                data_perity[f'{j}\{l}'] = data_df[j] / data_df[l]
```

```python
data_perity.drop(0, axis=1, inplace=True)
data_perity
```

```python
data_perity.to_excel('F:\my project\dataset\my pro-datasets\\data_perity.xlsx')
```
