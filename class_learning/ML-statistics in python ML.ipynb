```python
def inverse(ndarray: np.ndarray):
    inv = np.zeros((ndarray.shape[1], ndarray.shape[0]))
    assert type(ndarray) == np.ndarray
    for i in range(ndarray.shape[0] - 1):
        for j in range(ndarray.shape[1] - 1):
            inv[j, i] = ndarray[i, j]
    return inv


val2 = val[:, 1:]
val3 = inverse(val2)
val3
```

```python
def ECDF(data):
    n = len(data)
    x = np.sort(data)
    y = [i / n for i in range(n)]
    return x, y


inch = np.array([3, 5, 7, 5, 7, 9, 2, 4, 6, 3, 5, 9])

x, y = ECDF(inch)

print(len(x), len(y))
```

```python
plt.scatter(x, y)
plt.xlabel('INCH')
plt.ylabel('DISTRIBUTION')
```
