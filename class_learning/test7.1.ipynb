```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(size=100)
w = np.random.normal(size=100)
```

```python
for i in range(1, len(x)):
    x[i] = x[i - 1] + w[i]
```

```python
plt.plot(x)
```

```python
def moving_avrage(data: np.ndarray, length: int):
    avg = []
    j = 0
    for i in range(length, len(data)):
        avg.append(np.mean(data[j:i]))
        j += 1
    del j
    return np.array([np.nan]*(length)+avg)
```

```python
moving_avrage(x,20)
```

```python
plt.plot(x)
plt.plot(moving_avrage(x,10))
plt.plot(moving_avrage(x,20))
plt.plot(moving_avrage(x,30))
```
