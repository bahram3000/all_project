```python
df=pd.read_excel('F:\my project\dataset\population\\pop_.xlsx')
df
```

```python
df2=df.drop('کل',axis=1)
df2
```

```python
df_parity=pd.DataFrame(np.array([np.nan]*len(df2.index)).reshape(-1,1),index=df.index)
df_parity
```

```python

for i,j in enumerate(df2.columns):
    for k,l in enumerate(df2.columns):
        if j!=l:
            if k<i:
                #print(j)
                #print(l)
                df_parity[f'{j}\{l}']=df2[j]/df2[l]
```

```python
df_parity.drop(0,axis=1,inplace=True)
df_parity
```
