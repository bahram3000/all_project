```python
col=data.values[:,0]
ind=data.columns[1:]
```

```python
data=pd.DataFrame(data.values[:,1:].T,index=ind,columns=col)
data
```

```python
data.dropna(inplace=True)
data
```

```python
data['world_rate_bm']=np.nan
for i in range(len(data.index[1:])):
    data['world_rate_bm'][i]=data['World_bm_per_gdp'][i]/data['World_bm_per_gdp'][i-1]
data['iran_rate_bm']=np.nan
for i in range(len(data.index[1:])):
    data['iran_rate_bm'][i]=data['Iran, Islamic Rep_bm_per_gdp'][i]/data['Iran, Islamic Rep_bm_per_gdp'][i-1]
data
```

```python
data['world_inf']=1+data['World_inflation']/100
data['iran_inf']=1+data['Iran, Islamic Rep_inflation']/100
data
```

```python
data_final=data.loc[:,'world_rate_bm':]
data_final
```
