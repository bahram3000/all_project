```python
def equal_log(data1, data2):
    l = []
    if len(data1) > len(data2):
        for i in data1:
            for j in data2:
                if i == j:
                    l.append(True)
                    break
            else:
                l.append(False)
    else:
        for i in data2:
            for j in data1:
                if i == j:
                    l.append(True)
                    break
            else:
                l.append(False)
    return np.array(l)
```

```python
def intersect(data1, data2):
    l = []
    if len(data1) > len(data2):
        for i in data1:
            for j in data2:
                if i == j:
                    l.append(i)
                    break
    else:
        for i in data2:
            for j in data1:
                if i == j:
                    l.append(i)
                    break
    return np.array(l)
```
