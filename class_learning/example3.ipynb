```python
world_inf = data['world inflation'][17:41]
world_inf
```

```python
def inf_agg(data_1: pd.core.series.Series):
    list_ag = []
    ag = 1
    for i in data_1:
        ag = ag * (1 + i / 100)
        list_ag.append(ag)
    return list_ag


l = inf_agg(world_inf)
```

```python
l = pd.Series(l)
l
```

```python
mask = (data.columns != 'world inflation') & (data.columns != 'gold low') & (data.columns != 'gold high')

#data2 = data.drop(data.columns != ['world inflation', 'gold low', 'gold high'])
#data2
col = ['world inflation', 'gold low', 'gold high']
'gold' not in col
```

```python
def non_column_excepted(data_se, ex_column_list):
    non_ex = []
    for i in data_se.columns:
        if i not in ex_column_list:
            non_ex.append(i)
    return non_ex


non_col = non_column_excepted(data, col)
non_col
```

```python
data2 = data.drop(non_col, axis=1)
data2
```

```python
data2['gold avg'] = (data2['gold low'] + data2['gold high']) / 2
data2
```

```python
m = inf_agg(data2['world inflation'][17:])
data2['inflation aggregate'] = 0
for i in data2.index:
    if i < 17:
        data2['inflation aggregate'][i] = 0
    else:
        data2['inflation aggregate'][i] = m[i - 18]

data2
```

```python
data2['gold inf agg'] = data2['gold avg'][17] * data2['inflation aggregate']
data2
```

```python
data2['gold low inf adj'] = [0] * len(data2['gold low'])
for i in range(1, len(data2['gold low'])):
    data2['gold low inf adj'][i] = data2['gold low'][i - 1] * (1 + data2['world inflation'][i] / 100)

#data2['gold low inf adj'] = data2['gold low'] * (1 + data2['world inflation'] / 100)
data2['gold high inf adj'] = [0] * len(data2['gold high'])
for i in range(1, len(data2['gold high'])):
    data2['gold high inf adj'][i] = data2['gold high'][i - 1] * (1 + data2['world inflation'][i] / 100)
#data2['gold high inf adj'] = data2['gold high'] * (1 + data2['world inflation'] / 100)
#data2['gold avg inf adj'] = data2['gold avg'] * (1 + data2['world inflation'] / 100)
data2['gold avg inf adj'] = [0] * len(data2['gold avg'])
for i in range(1, len(data2['gold high'])):
    data2['gold avg inf adj'][i] = data2['gold avg'][i - 1] * (1 + data2['world inflation'][i] / 100)

data2
```
