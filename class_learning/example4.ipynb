```python
data = pd.DataFrame(val, index=year, columns=col)
data = data.drop([1402], axis=0)
data
```

```python
def inf_agg(data_1: pd.core.series.Series):
    list_ag = [1] * (len(data_1) + 1)
    ag = 1
    j = 1
    for i in data_1.index:
        ag = ag * (1 + (data_1[i] / 100))
        list_ag[j] = ag
        j += 1
    return list_ag


l_agg = inf_agg(data['infaltion'])
l_agg
```

```python
data['avrage'] = (data.low + data.high) / 2
data
```

```python
data['infaltion_aggregate'] = l_agg[:-1]
data
```

```python
def gdp_growth_agg(data_1: pd.core.series.Series):
    list_ag = [1] * (len(data_1) + 1)
    ag = 1
    j = 1
    for i in data_1.index:
        ag = ag * ((100 + data_1[i]) / 100)
        list_ag[j] = ag
        j += 1
    return list_ag


gl_agg = gdp_growth_agg(data['gdp growth'])
gl_agg
```

```python
data['gdp_growth_aggregate'] = gl_agg[:-1]
data
```

```python
min_=np.min(data.index)
data['low_inf_gdp_adj'] = len(data.index) * data.low[min_]

for i in data.low.index[1:]:
    data.low_inf_gdp_adj[i] = data.low[i-1] * (1 + (data.infaltion[i]/100)) * ((100 + data['gdp growth'][i])/100)

data['high_inf_gdp_adj'] = len(data.index) * data.high[min_]
for i in data.high.index[1:]:
    data.high_inf_gdp_adj[i] = data.high[i-1] * (1 + (data.infaltion[i]/100)) * ((100 + data['gdp growth'][i])/100)
data['avg_inf_gdp_adj'] = len(data.index) * data.avrage[min_]
for i in data.avrage.index[1:]:
    data.avg_inf_gdp_adj[i] = data.avrage[i-1] * (1 + (data.infaltion[i]/100)) * ((100 + data['gdp growth'][i])/100)
data
```

```python
data['ROR']=data.avrage/data.avg_inf_gdp_adj
data
```

```python
data['inf_gdp_Stock']=data.infaltion_aggregate*data.gdp_growth_aggregate*data.avrage[1371]
data
```
