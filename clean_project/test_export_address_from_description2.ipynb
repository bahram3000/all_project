```python
persian_chars = 'حکشخذدآرقتضظفگموصطزهثسژبجعپیلنا'
```

```python
def preproces_persian_text(inp_str: str):
    out = ''
    for i in inp_str:
        if i in persian_chars:
            out += i
        else:
            out += ' '
    return out
```
