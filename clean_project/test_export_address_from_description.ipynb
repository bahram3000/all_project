```python
def remove_all_symbols(inp_srt:str):
    out=inp_srt
    symb='''!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~،'''
    for i in symb:
        out=out.replace(i,'')
    spaces=['\n','\t','\r']
    for i in spaces:
        out=out.replace(i,' ')
    return out
```

```python
def persian_letters(inp_str:str):
    out=inp_str
    english_letters='abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789'
    for i in english_letters:
        out=out.replace(i,' ')
    return out
```
