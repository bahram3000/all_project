```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
iris=load_iris()
x=iris.data
kmn=KMeans(n_clusters=3)
kmn.fit(x)
labels=kmn.predict(x)
labels
```

```python
center=kmn.cluster_centers_
center
```

```python
plt.scatter(x[:,0],x[:,2],c=labels)
plt.scatter(center[:,0],center[:,2],marker='*',s=150,c='red')
```

```python
kmn.inertia_
```

```python
score=[]
for j in range(1,20):
    kmn2=KMeans(n_clusters=j)
    kmn2.fit(x)
    #labels2=kmn2.predict(x)
    print(j,' : ',kmn2.inertia_)
    score.append(kmn2.inertia_)
```

```python
plt.plot(score)
plt.xlabel('n_cluster')
plt.ylabel('score')
```

```python
from scipy.cluster.hierarchy import linkage,dendrogram,fcluster
```

```python
hirarical=linkage(x,method='complete')
dendrogram(hirarical)
```

```python
labels=fcluster(hirarical,4,criterion='distance')
labels
```

```python
from sklearn.cluster import MeanShift
```

```python
m_s=MeanShift()
m_s.fit(x)
#labels=m_s.predict(x) #or labels =m_s.labels_
labels2=m_s.labels_
center2=m_s.cluster_centers_
```

```python
plt.scatter(x[:,0],x[:,2],c=labels2)
plt.scatter(center2[:,0],center2[:,2],marker='*',c='red',s=150)
```

```python
m_s2=MeanShift(bandwidth=0.5)
m_s2.fit(x)
labels3=m_s2.predict(x) #or labels =m_s.labels_
#labels2=m_s.labels_
center3=m_s2.cluster_centers_
plt.scatter(x[:,0],x[:,2],c=labels3)
plt.scatter(center3[:,0],center3[:,2],marker='*',c='red',s=50)
```

```python
from sklearn.cluster import DBSCAN
```
