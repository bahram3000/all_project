```python
from sklearn.datasets import load_breast_cancer
bcd=load_breast_cancer()
x=bcd.data
y=bcd.target
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
x_train,x_tset,y_train,y_tset=train_test_split(x,y,test_size=0.3,shuffle=True)
```

```python
tree=DecisionTreeClassifier()
rand_p={'max_depth':[None,1,3],'max_features':randint(1,9),'min_samples_leaf':randint(1,9)}
tree_cv=RandomizedSearchCV(tree,rand_p,cv=10)
tree_cv.fit(x_train,y_train)
```

```python
tree_cv.best_params_
```

```python
tree_cv.best_score_
```

```python
tree_cv.score(x_tset,y_tset)
```
