```python
from math import floor
```

```python
list_score_by_train_size = []
for i in range(2, 100):
    tr_sz = i / 100
    knn3 = KNeighborsClassifier(n_neighbors=2+i//10)
    x_train1, x_test1, y_train1, y_test1 = train_test_split(x1, y, train_size=tr_sz, shuffle=True)
    knn3.fit(x_train1,y_train1)
    s=knn3.score(x_test1,y_test1)
    list_score_by_train_size.append(s)
```

```python
list_score_by_train_size=np.array(list_score_by_train_size)
plt.plot(list_score_by_train_size)
```

```python
list_score_by_train_size.max()
```

```python
def find_ind(data_1, element):
    j = 0
    for i in data_1:
        if i == data_1.max():
            return j
        j += 1
```

```python
where_max=find_ind(list_score_by_train_size,list_score_by_train_size.max())
where_max
```

```python
list_score_by_k_n=[]
tr_sz2=where_max/100
for i in range(2,len(iris_data.index)):
    if i < tr_sz2*len(iris_data.index):
        knn4=KNeighborsClassifier(n_neighbors=i)
        x_train2,x_test2,y_train2,y_test2=train_test_split(x1,y,train_size=tr_sz2,shuffle=True)
        knn4.fit(x_train2,y_train2)
        s=knn4.score(x_test2,y_test2)
        list_score_by_k_n.append(s)
```

```python
list_score_by_k_n=np.array(list_score_by_k_n)
plt.plot(list_score_by_k_n)
```
