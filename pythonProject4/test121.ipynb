```python
from pycoingecko import CoinGeckoAPI
import pandas as pd
```

```python
cg=CoinGeckoAPI()
```

```python
bitcoin_data=cg.get_coin_market_chart_by_id(id='bitcoin',vs_currency='usd',days=30)
bitcoin_data
```

```python
bitcoin_data['prices']
```

```python
bitcoin_df=pd.DataFrame(bitcoin_data['prices'],columns=['Timestamp','price'])
bitcoin_df
```

```python
bitcoin_df['Date']=pd.to_datetime(bitcoin_df['Timestamp'])
bitcoin_df
```

```python
candlestick_df=bitcoin_df.groupby(bitcoin_df.Date.dt.date).agg({'price':['min','max','first','last']})
```
