```python
def distance(data_1, data_2):
    assert len(data_1) == len(data_2)
    dist = 0
    for i in range(len(data_1)):
        dist += (data_1[i] - data_2[i]) ** 2
    return dist ** 0.5
```

```python
def find_near(data_1, intended_1):
    lis_dist = []
    for i in data_1.index:
        dist = distance(data_1.loc[i][:], intended_1)
        lis_dist.append(dist)
    lis_dist = np.array(lis_dist)
    min = lis_dist.min()
    for i in range(len(lis_dist)):
        if lis_dist[i] == min:
            return i
```

```python
def train_test_slicer(data_1, train_size: float = 0.7):
    ln = len(car_data2.index)
    size_train = math.floor(train_size * ln)
    ind_all = [random.randint(0, ln) for _ in range(ln)]
    ind_train = ind_all[:size_train]
    ind_train = pd.Series(ind_train)
    ind_test = ind_all[size_train:]
    ind_test = pd.Series(ind_test)
    col_1 = car_data2.columns
    c_data_train = pd.DataFrame(np.zeros((size_train, len(col_1))), index=ind_train, columns=col_1)
    for i in col_1:
        for j in ind_train:
            if j != 26 and j != 62:
                c_data_train[i][j] = car_data2[i][j]
    size_test = ln - size_train
    c_data_test = pd.DataFrame(np.zeros((len(ind_test), len(col_1))), index=ind_test, columns=col_1)
    for i in col_1:
        for j in ind_test:
            if j != 26 and j != 62:
                c_data_test[i][j] = car_data2[i][j]
    return c_data_train, c_data_test
```

```python
def score(data_1):
    true_num = 0
    all_1 = 0

    for i in data_1.index:
        all_1 += 1
        a_3 = data_1.loc[i][:]
        a_3.drop(['num_of_door', 'body_num'], axis=0, inplace=True)
        res = find_near(gr_cr, a_3)
        real = data_1.loc[i]['body_num']
        if res == real:
            true_num += 1
    return true_num / all_1
```
