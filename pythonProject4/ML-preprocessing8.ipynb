```python
def change_ind_col(data_1, axis):
    dic_ch = {}
    if axis == 0:
        lnx = len(data_1.index)
        ind = data_1.index
        for i in range(lnx):
            dic_ch[ind[i]] = str(i)
        data_1.rename(index=dic_ch, inplace=True)
    if axis == 1:
        lny = len(data_1.columns)
        ind = data_1.columns
        for i in range(lny):
            dic_ch[ind[i]] = str(i)
        data_1.rename(columns=dic_ch, inplace=True)
```

```python
def get_col_exc(data_1, list_exc: list):
    col = data_1.columns
    col = list(col)
    l_n = []
    for i in col:
        if i not in list_exc:
            l_n.append(i)
    return l_n
```

```python
def standard_data(data_1):
    ind = data_1.index
    col = data_1.columns
    val = data_1.values
    c_data = pd.DataFrame(val, index=ind, columns=col)
    for i in c_data.columns:
        avm = c_data[i].mean()
        st = c_data[i].std()
        c_data[i] = (c_data[i] - avm) / st
    return c_data
```
