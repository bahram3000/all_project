```python
persian_number = ['۰', '۱', '۲', '۳', '۴', '۵', '۶', '۷', '۸', '۹']
```

```python
per_to_eng_dict = {'۰': '0', '۱': '1', '۲': '2', '۳': '3', '۴': '4', '۵': '5', '۶': '6', '۷': '7', '۸': '8', '۹': '9'}
```

```python
def extract_persian_num_from_string(text: str):
    out = ''
    for i in text:
        if i in persian_number:
            out += per_to_eng_dict[i]
    if out != '':
        return int(out)
    else:
        np.NaN
```

```python
l = []
for i in b.values:
    a = []
    for j in i:
        a.append(extract_persian_num_from_string(j))
    l.append(a)
```
