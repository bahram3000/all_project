```python
last_date = 1699934480450773
b = []
for i in range(30):
    if i != 0:
        js1 = {"json_schema": {
            "category": {
                "value": "apartment-rent"
            },
            "query": "اجاره اپارتمان",
            "sort": {
                "value": "sort_date"
            },
            "cities": [
                "30"
            ]},
            "last-post-date": last_date
        }
        headers1 = {
            "Content_type": "application/json"
        }
        response = requests.post(url=url, json=js, headers=headers)
        data1 = response.json()
        for i in data1['web_widgets']['post_list']:
            b.append([i['data']['title'], i['data']['top_description_text'], i['data']['middle_description_text']])
        last_date = data1['last_post_date']
```

```python
b = np.array(b)
b
```

```python
persian_number = ['۰', '۱', '۲', '۳', '۴', '۵', '۶', '۷', '۸', '۹']
```

```python
per_to_eng_dict = {'۰': '0', '۱': '1', '۲': '2', '۳': '3', '۴': '4', '۵': '5', '۶': '6', '۷': '7', '۸': '8', '۹': '9'}
```

```python
def extract_persian_num_from_string(text: str):
    out = ''
    for i in text:
        if i in persian_number:
            out += per_to_eng_dict[i]
    if out != '':
        return int(out)
    else:
        np.NaN
```

```python
extract_persian_num_from_string(b[4, 1])
```

```python
extract_persian_num_from_string(b[2, 0])
```

```python
l = []
for i in b:
    a = []
    for j in i:
        a.append(extract_persian_num_from_string(j))
    l.append(a)
```

```python
l=np.array(l)
l
```

```python
l=pd.DataFrame(l,columns=['meter','base','rent'])
l
```

```python
l.dropna(inplace=True)
l
```

```python
l['all_rent_per_m']=(l['base']*0.03+l['rent'])/l['meter']
l
```
