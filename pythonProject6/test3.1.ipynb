```python
class time_series():
    def __int__(self, X):
        self.tme_ser = X

    def f_x(self, x, b):
        a = np.max(x)
        return a * math.sin(b * x)

    def f_l(self, b):
        a = np.max(self.tme_ser)
        c = np.min(self.tme_ser)
        return [math.sin(i*b)+1+c for i in range(len(self.tme_ser))]

    @staticmethod
    def errors(X, Y):
        if len(X) == len(Y):
            return [X[i] - Y[i] for i in range(len(X))]

    @staticmethod
    def mse(errors_):
        mse = 0.0
        for i in errors_:
            mse += i ** 2
        return mse / len(errors_)

    def train(self, walk, iter):
        b = np.random.rand()
        pred = (self.f_l(b))
        mse1 = self.mse(self.errors(pred, self.tme_ser))
        mse2 = self.mse(self.errors(self.f_l(b + 1 * walk), self.tme_ser))
        for i in range(1, iter):
            if mse2 > mse1:
                b = b - i * walk
                pred = (self.f_l(b))
                mse1 = mse2
                mse2 = self.mse(self.errors(pred, self.tme_ser))
            elif mse2 < mse1:
                b = b + i * walk
                pred = (self.f_l(b))
                mse1 = mse2
                mse2 = self.mse(self.errors(pred, self.tme_ser))
        return self.f_l(b),b

```
