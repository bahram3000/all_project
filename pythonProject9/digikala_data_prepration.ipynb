```python
final_data = []
index_dt = []
for i in image_files_path:
    try:
        im_v = Image.open(up_path + '\\' + i)
        im_v = im_v.resize((100, 100))
        im_v = np.array(im_v)
        final_data.append(im_v.reshape((1, 30000)))
        index_dt.append(i)
    except:
        print(i, end=' ')
        #exept.append(i)
```

```python
final_data = np.array(final_data)
final_data
```

```python
final_data_df = pd.DataFrame(final_data, index=index_dt)
final_data_df
```

```python
rate = pd.read_excel(up_path + '\\' + rate_path)
rate.rename(columns={'Unnamed: 0': 'image_name'}, inplace=True)
rate.set_index('image_name', inplace=True)
rate
```
