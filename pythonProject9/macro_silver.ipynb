```python
def intersect(dt1, dt2):
    intersect_list = []
    if len(dt1) >= len(dt2):
        for i in dt1:
            #print(i)
            if i in dt2:
                intersect_list.append(i)
    else:
        for i in dt2:
            #print(i)
            if i in dt2:
                intersect_list.append(i)
    return intersect_list
```

```python
def blind_moving_average(dt1):
    return np.array([np.mean(dt1[:i + 1]) for i in range(len(dt1))])
```

```python
blind_moving_average(google_usege[int_net_dom_google_use])
```

```python
def blind_moving_average_bar(dt1, bar: int):
    a = blind_moving_average(dt1)
    for i in range(bar - 1):
        a = blind_moving_average(a)
    return a
```

```python
blind_moving_average_bar(google_usege[int_net_dom_google_use], 18)
```

```python
plt.plot(google_usege[int_net_dom_google_use].values)
plt.plot(blind_moving_average(google_usege[int_net_dom_google_use].values))
plt.plot(blind_moving_average_bar(google_usege[int_net_dom_google_use], 2))
plt.plot(blind_moving_average_bar(google_usege[int_net_dom_google_use], 3))
#plt.plot(blind_moving_average_bar(google_usege[int_net_dom_google_use],18))
```
